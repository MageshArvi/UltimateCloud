```python
windowfn_df = spark.read.format("csv").option("header",True).option("inferSchema",True).load('abfss://unitycatalog@databricksstudysa.dfs.core.windows.net/catalog/windowdatamodified.csv')
```

```python
windowfn_df.show()
```

1. Lead Compares it with the next row
2. Lag Compares it with the previous row

```python
from pyspark.sql.functions import lead, lag
from pyspark.sql.window import Window
```

```python
mywin = Window.partitionBy("country").orderBy("weeknum")
```

```python
lag_df = windowfn_df.withColumn("previous_week",lag("invoicevalue").over(mywin))
```

```python
lag_df.show()
```

```python
from pyspark.sql.functions import expr
```

```python
final_df = lag_df.withColumn("invoice_diff",expr("invoicevalue - previous_week"))
```

```python
final_df.show()
```
