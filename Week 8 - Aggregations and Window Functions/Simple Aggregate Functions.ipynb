Aggregation means crunching some of the records to get some ouput rows.

We have 3 Aggregations.

1. Simple Aggregations - Gives only one output row

    - count the total number of records
    - find the sum of quantities

2. Grouping Aggregations - Will do a group by

3. Windowing Aggregations

```python
ordersdata_df = spark.read.format("csv").option("header",True).option("inferSchema",True).load('abfss://unitycatalog@databricksstudysa.dfs.core.windows.net/catalog/order_data.csv')
```

```python
ordersdata_df.show()
```

Let's do simple aggregation.
How many ever records the dataset has, we get the final output as single record

1. Let's count the total number of records
2. Count the number of distinct invoice IDs
3. Sum of quantities
4. Avg Unit Price

We can achieve all these using 3 different styles

    - Programmatic Style
    - Column Expression
    - Spark SQL Style

```python
from pyspark.sql.functions import *
```

```python
ordersdata_df.select(count("*").alias("row_count"), countDistinct("InvoiceNo").alias("unique_invoice"),sum("quantity").alias("sum_quantities"),avg("unitprice").alias("avg_unit_price")).show()
```

```python
ordersdata_df.selectExpr("count(*) as row_count","count(distinct(invoiceno)) as unique_invoice","sum(quantity) as sum_quantities","avg(unitprice) as avg_unit_price").show()
```

```python
ordersdata_df.createOrReplaceTempView("orders")
```

```python
spark.sql("select count(*) as row_count, count(distinct(invoiceno)) as unique_invoice, sum(quantity) as sum_quantities, avg(unitprice) as avg_unit_price from orders").show()
```
