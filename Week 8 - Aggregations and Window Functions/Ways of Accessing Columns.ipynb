```python
orders_schema = 'order_id long, order_date date, customer_id long, order_status string'
```

```python
orders_df = spark.read.format("csv").schema(orders_schema).load('abfss://unitycatalog@databricksstudysa.dfs.core.windows.net/catalog/orders_1gb.csv')
```

```python
orders_df.select("*").show()
```

```python
orders_df.select("order_id","order_status").show()
```

```python
from pyspark.sql.functions import *
```

```python
orders_df.select("order_id", orders_df.order_status,orders_df['order_date'],column('customer_id'),col('customer_id'),expr("order_status")).show()
```

```python
orders_df.select("order_id","customer_id",expr("customer_id + 1 as new_cust_id")).show()
```

```python
orders_df.select("order_id","order_status").filter(col("order_status").like("PENDING%")).show()
```

```python
orders_df.select("order_id","order_status").where(col("order_status").like("PENDING%")).show()
```
