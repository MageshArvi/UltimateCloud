```python
orders_df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
        .load("dbfs:/FileStore/shared_uploads/talk2mageshacc@outlook.com/orders_wh.csv")
```

```python
orders_df.printSchema()
```

```python
orders_df.show()
```

```python
orders_df.createOrReplaceTempView("orders")
```

1. Top 15 customers who placed the most number of orders

```python
spark.sql("select customer_id, count(*) as total_orders from orders group by customer_id order by total_orders desc limit 15").show()
```

```python
df = orders_df.groupBy("customer_id").count().sort("count",ascending=False).limit(15)
```

```python
df.show()
```

2. Find number of orders under each order status

```python
df_2 = orders_df.groupby("order_status").count()
```

```python
df_2.show()
```

```python
spark.sql("select order_status, count(*) from orders group by order_status").show()
```

3. No of active customers

```python
spark.sql("""select count(*) from (select customer_id,count(*) 
          from orders
          group by customer_id
          having count(*) >= 1)""").show()
```

```python
spark.sql("select count(distinct customer_id) from orders").show()
```

```python
results = orders_df.select("customer_id").distinct().count()
```

4. Customers with most number of closed orders

```python
spark.sql("select customer_id, count(*) as total_orders  from orders where order_status = 'CLOSED' group by customer_id order by total_orders desc" ).show()
```

```python
results_df_count = orders_df.filter("order_status = 'CLOSED'").groupBy("customer_id").count().sort("count",ascending=False)
```

```python
results_df_count.show()
```
