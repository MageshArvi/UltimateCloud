```python
from pyspark.sql.functions import *
```

```python
ordersdata_df = spark.read.format("csv").option("header",True).option("inferSchema",True).load('abfss://unitycatalog@databricksstudysa.dfs.core.windows.net/catalog/order_data.csv')
```

1. Total Quantities based on Invoice Number and Country
2. Invoice Value based on Invoice Number and Country

```python
summary_df = ordersdata_df.groupBy("country","invoiceno") \
    .agg(sum("quantity").alias("total_quantity"), sum(expr("quantity * unitprice")).alias("invoice_value")).sort("invoiceno")
```

```python
summary_df.show()
```

```python
summary_df_1 = ordersdata_df.groupby("country","invoiceno") \
    .agg(expr("sum(quantity) as total_quantity"), expr("sum(quantity*unitprice) as invoice_value")).sort("invoiceno")
```

```python
summary_df_1.show()
```

```python
ordersdata_df.createOrReplaceTempView("ordersdata")
```

```python
spark.sql("select country, invoiceno,sum(quantity) as total_quantity, sum(unitprice * quantity) as total_value from ordersdata group by country,invoiceno order by invoiceno").show()
```
