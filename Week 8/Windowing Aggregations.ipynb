```python
windows_df = spark.read.format("csv").option("header",True).option("inferSchema",True).load('abfss://unitycatalog@databricksstudysa.dfs.core.windows.net/catalog/windowdata.csv')
```

```python
windows_df.show()
```

```python
windows_df.sort('country').show()
```

1. Partition based on country
2. Sort based on week num
3. Window size

```python
windows_df.printSchema()
```

```python
from pyspark.sql import *
```

```python
mywindow = Window.partitionBy("country").orderBy("weeknum").rowsBetween(Window.unboundedPreceding,Window.currentRow)
```

```python
from pyspark.sql.functions import sum
```

```python
result_df = windows_df.withColumn("running_total",sum("invoicevalue").over(mywindow))

```

```python
result_df.show()
```

```python
my_window = Window.partitionBy("country").orderBy("weeknum").rowsBetween(-2,Window.currentRow)
```

```python
result_final_df = windows_df.withColumn("running_total",sum("invoicevalue").over(my_window))

```

```python
result_final_df.show()
```
