```python
orders_schema = 'order_id long, order_date date, customer_id long, order_status string'
```

```python
orderfile_df = spark.read.format("csv").schema(orders_schema).load('abfss://unitycatalog@databricksstudysa.dfs.core.windows.net/catalog/orders_1gb.csv')
```

```python
spark.sql("create database cachingdemo_db")
```

```python
spark.sql("use cachingdemo_db")
```

```python
orderfile_df.write.format("delta").saveAsTable("cachingdemo_db.orders")
```

```python
spark.sql("describe formatted cachingdemo_db.orders").show()
```

```python
spark.sql("select count(*) from cachingdemo_db.orders").show()
```
