```python
orders_schema = 'order_id long, order_date date, customer_id long, order_status string'
```

```python
orders_data_df = spark.read.format("csv").schema(orders_schema).load('abfss://unitycatalog@databricksstudysa.dfs.core.windows.net/catalog/orders_1gb.csv')
```

```python
orders_data_df.show()
```

```python
orders_data_df.write.saveAsTable('cachingdemo_db.orders_data')
```

```python
spark.catalog.currentDatabase()
```

```python
spark.sql("clear cache")
```

```python
spark.sql("cache table cachingdemo_db.orders_data")
```
