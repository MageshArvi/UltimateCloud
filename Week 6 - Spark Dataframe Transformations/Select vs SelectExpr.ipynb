```python
products_df = spark.read.csv("dbfs:/FileStore/datasets/Products.csv",header=True, inferSchema = True)
```

```python
products_df.show()
```

```python
products_df = products_df.withColumnRenamed("Product Name","Product_Name") \
    .withColumnRenamed("ProductKey","ProductKey") \
    .withColumnRenamed("Unit Cost USD","Unit_Cost_USD") \
    .withColumnRenamed("Unit Price USD","Unit_Price_USD") 
```

```python
products_df.show()
```

```python
new_products_df = products_df.withColumn("Unit_Cost_UD", expr("CASE WHEN Product_Name like '%Contoso%' THEN Unit_Cost_USD * 1.2 WHEN Product_Name like '%Adventure%' THEN Unit_Cost_USD * 1.1 ELSE Unit_Cost_USD END"))
```

```python
new_products_df.show()
```
